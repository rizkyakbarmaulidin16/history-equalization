```python
# Import library yang dibutuhkan
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from google.colab import files

# Fungsi untuk Histogram Equalization
def histogram_equalization(image_array):
    # Hitung histogram
    hist, bins = np.histogram(image_array.flatten(), bins=256, range=[0, 256])

    # Hitung CDF (Cumulative Distribution Function)
    cdf = hist.cumsum()
    cdf_min = cdf[cdf > 0].min()
    cdf_normalized = (cdf - cdf_min) / (cdf.max() - cdf_min) * 255
    cdf_normalized = cdf_normalized.astype('uint8')

    # Terapkan transformasi ke gambar
    equalized_image = cdf_normalized[image_array]
    return equalized_image

# Unggah file gambar
uploaded = files.upload()

# Proses file yang diunggah
for filename in uploaded.keys():
    # Membaca gambar
    img = Image.open(filename).convert('L')  # Ubah ke grayscale
    img_array = np.array(img)

    # Lakukan Histogram Equalization
    equalized_img_array = histogram_equalization(img_array)
    equalized_img = Image.fromarray(equalized_img_array)

    # Menampilkan hasil
    plt.figure(figsize=(12, 6))

    # Gambar asli
    plt.subplot(1, 3, 1)
    plt.title("Gambar Asli")
    plt.imshow(img_array, cmap='gray')
    plt.axis('off')

    # Histogram asli
    plt.subplot(1, 3, 2)
    plt.title("Histogram Asli")
    plt.hist(img_array.flatten(), bins=256, color='blue', alpha=0.7)

    # Gambar setelah Histogram Equalization
    plt.subplot(1, 3, 3)
    plt.title("Gambar Setelah Equalization")
    plt.imshow(equalized_img_array, cmap='gray')
    plt.axis('off')

    plt.show()

    # Simpan gambar hasil
    output_filename = "equalized_" + filename
    equalized_img.save(output_filename)
    print(f"Gambar hasil disimpan sebagai {output_filename}")
```
